## 3. 명사 출현 빈도 세기

```python
from collections import Counter
```

```python
# 테스트로 5개 글에 대해서 실행
word_counter = Counter()
twitter = Twitter()


for content in contents[:5]:
    nouns = twitter.nouns(content)
    word_counter.update(nouns)
```

```python
word_counter.most_common(100)
```

```python
# 글 전체 분석

word_counter = Counter()
twitter = Twitter()

for content in contents:
    nouns = twitter.nouns(content)    # 명사만 골라서 리스트 형태로 리턴해줌
    word_counter.update(nouns)        #  
    
```

```python
word_counter.most_common(100)
```

```python
len(word_counter)
```

```python
counters = pd.DataFrame(word_counter.most_common(26802))
counters.columns = ['단어','빈도수']
```

```python
counters.head()
```

```python
counters.to_csv('word_count.csv', encoding = 'cp949')
```
