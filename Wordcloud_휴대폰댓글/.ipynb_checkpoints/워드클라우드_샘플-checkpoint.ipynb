```python
import nltk
from konlpy.tag import Twitter
from matplotlib import font_manager, rc
import platform
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
```

```python
#텍스트 데이터 읽기
with open('비트코인댓글.txt','r') as inputfile:
    data=inputfile.read()
print(data)


#csv 파일일경우
'''
with open("sample.csv") as f:
    lines = f.readlines() 
data = "".join(lines)
'''
```

```python
# 글꼴 지정

if platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
elif platform.system()=='Windows' :
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:
    print('unknown system')
```

```python
#빈도수 정리하기

stopwords=['년','놈','샴숑']

t=Twitter()
nouns=t.nouns(data)  # 명사만 분류하기
counts = Counter(nouns)
noun_lists=counts.most_common(200)   #적당수로 계산함
refined_nouns=[]

#금지어 제외함
for noun in noun_lists:
    if noun[0] in stopwords:
        continue
    refined_nouns.append(noun)

print(refined_nouns)
```

```python
#워드클라우드(블랙)
%matplotlib inline

tmp = dict(refined_nouns)
wordcloud=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf", relative_scaling=0.3, 
                    background_color="black", max_words=50,
                   font_step=10).generate_from_frequencies(tmp)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')

#파일명 확인 필요
plt.savefig('sample(비트코인)_black.png')
```

```python
#워드클라우드(화이트)
%matplotlib inline


tmp = dict(refined_nouns)
wordcloud=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf", relative_scaling=0.3, 
                    background_color="white", max_words=50,
                   font_step=10).generate_from_frequencies(tmp)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')

#저장파일명 확인
plt.savefig('sample(비트코인)_white.png')
```

```python
#마스크 준비
from wordcloud import ImageColorGenerator

#이미지 --> 행렬로 변환
symbol_mask = np.array(Image.open('./symbol/photo.jpg'))

#이미지 위치별 색상 따오기
image_colors = ImageColorGenerator(symbol_mask)
```

```python
#워드클라우드(심볼)

tmp = dict(refined_nouns)
wordcloud=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf", relative_scaling=0.3, 
                    background_color="white", max_words=100,
                   mask = symbol_mask,font_step=20).generate_from_frequencies(tmp)
plt.figure(figsize=(16,8))
# plt.imshow(wordcloud, interpolation="bilinear")
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")    #심볼 색상 활용
plt.axis('off')
plt.savefig('심볼모양(비트코인).png')
```
