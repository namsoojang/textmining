```python
import nltk
from konlpy.tag import Twitter
from matplotlib import font_manager, rc
import platform
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
with open('data.txt','r') as inputfile:
    data=inputfile.read()
# print(data)
```

```python
if platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
elif platform.system()=='Windows' :
    fornt_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:
    print('unknown system')
```

```python
%matplotlib inline

t=Twitter()
nouns=t.nouns(data)  #  

ko=nltk.Text(nouns,name="")   
ranking=ko.vocab().most_common(100)   #  10
tmp = dict(ranking)
# print(tmp)
wordcloud=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf", relative_scaling=0.3, background_color="white").generate_from_frequencies(tmp)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('test2 .png')
```
