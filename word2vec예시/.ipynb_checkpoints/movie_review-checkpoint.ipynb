```python
f = open('token_train.txt','rb')
token_train = pickle.load(f)
token_train
```

```python
f = open('token_test.txt','rb')
token_test = pickle.load(f)
token_test
```

```python
def pumsa(doc):    
    token = []
    rate = []
    
    for items in doc:
        rate.append(items[1])
        
        words = []
        for item in items[0]: 
            if item.split('/')[1] in ['Noun', 'Verb', 'Adjective']:
                words.append(item)
        token.append(words)
    
    return token, rate
```

```python
pumsa(token_train)
```

```python
train_review, train_rate = pumsa(token_train)
```

```python
train_review
```

```python
train_rate
```

```python
test_review, test_rate = pumsa(token_test)
```

```python
stop_words = ['하다', '되다', '이', '것', '점', '이다', '보고', '거', '때', '그', '수', '이다', '나다', '난',
             '걸', '듯', '함', '임', '전', '도', '게', '저', '요', '돼다', '기', '데', '감', '스럽다', '제']
```

```python
# stop words 제거 함수
def rm_stop(token):
    final = []
    for words in token:
        word_list = []
        for word in words:
            if word.split("/")[0] not in stop_words:
                word_list.append(word)
        final.append(word_list)
    return final
```

```python
train_review = rm_stop(train_review)
```

```python
test_review = rm_stop(test_review)
```

# 2. word2vec

```python
# word2vec model for train set
from gensim.models import Word2Vec
model = Word2Vec(train_review, size=100, window=10, min_count=10, workers=4)
```

```python
# word2vec embedding vectors for words
w2v = dict(zip(model.wv.index2word, model.wv.syn0))
```

# 2-1. 문서 embedding

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import defaultdict
```

```python
# word2vec embedding vector와 word에 대한 tfidf 가중치를 이용한 vectorizer 함수
class TfidfEmbeddingVectorizer:
    def __init__(self, word2vec):
        self.word2vec = word2vec
        
    def transform(self, X):
        tfidf = TfidfVectorizer(analyzer = lambda x : x) 
        tfidf.fit(X)
        max_idf = max(tfidf.idf_) 
        word2weight = defaultdict(lambda : max_idf, [(w, tfidf.idf_[i]) for w, i in tfidf.vocabulary_.items()]) 
        
        array_list =[]
        for words in X:
            array_list.append(np.array(np.mean([self.word2vec[w]*word2weight[w] for w in words if w in self.word2vec] or [np.zeros(100)], axis = 0)))
        return(array_list)
```

```python
vec_tf = TfidfEmbeddingVectorizer(w2v)
```

```python
train_tf = vec_tf.transform(train_review)
```

```python
test_tf = vec_tf.transform(test_review)
```

# 3. 분석 - SVM

```python
# trainset에 대한 rbf-kernel SVC model
from sklearn.svm import SVC
clf = SVC(decision_function_shape='ovo')
svc_clf = clf.fit(train_tf, train_rate)
```

```python
pred = svc_clf.predict(test_tf)
pred
```

```python
from sklearn import metrics
print(metrics.classification_report(test_rate, pred))
```
