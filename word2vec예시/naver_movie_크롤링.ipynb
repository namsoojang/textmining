```python
import re
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
def get_movie(page_num):
    session = requests.Session()
    headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_5) AppleWebKit537.36 (KHTML, like Gecko) Chrome", "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8"}

    page_url = "http://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=152170&target=after&page={}".format(page_num)
    r = session.get(page_url, headers=headers)
    soup = BeautifulSoup(r.text, 'html.parser')
    
    trs = soup.findAll('table')[1].tbody.findAll('tr')
    
    total = []
    for tr in trs:
        point = tr.find('td', {'class':'point'}).text
        title = tr.find('td', {'class':'title'}).find("br").text.replace("\n","").replace("\t","").replace("\r","")
        
        total.append([point, title])
    
    return pd.DataFrame(total, columns=['point', 'title'])
```

```python
get_movie(1)
```

```python
def get_all_movie(page_range):
    movie = pd.DataFrame()
    
    for page_num in page_range:
        sub_movie = get_movie(page_num)
        movie = movie.append(sub_movie, ignore_index=True)

    return movie
```

```python
KIN = get_all_movie(range(1,501))
```
