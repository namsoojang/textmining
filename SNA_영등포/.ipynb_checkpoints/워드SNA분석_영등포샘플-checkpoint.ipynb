```python
from konlpy.tag import Komoran
```

```python
#데이터 불러오기.  한줄에 댓글 하나 → 리스트로 댓글별 저장
INPUT_FILE = './data/토론샘플.txt'

datas=[]
with open(INPUT_FILE, "r") as input_file:
    for line in input_file:
        if len(line)==0:
            continue
        datas.append(line.strip())
#     data=input_file.read()

print(datas)        
```

```python
# 댓글별 형태소 분석하기

ma_results = []
komoran = Komoran()

for line in datas:
    ma_by_line=komoran.pos(line)
    ma_results.append(ma_by_line)

print(ma_results)
```

```python
print(len(ma_results)) #233  233개의 댓글
```

```python
print(ma_results[3])
print(type(ma_results[3]))
```

```python
# 댓글별 원하는 품사만 저장하기

FILTER_IN_POSES=['NNG','NNP']   
FILTER_IN_POSES_AD1=['XR']     #어근.~하다 ex) 유쾌  
FILTER_IN_POSES_AD2=['VA']      #형용사. ~다 ex) 쉽, 어렵
#NNG 일반명사  NNP 고유명사 XR 어근 (~하다) VA 형용사(~다)

ma_results_selected=[]

for ma_result in ma_results:
    ma_selected=[]
    for morph_lex, morph_cat in ma_result:
        if morph_cat in FILTER_IN_POSES:   # 일반명사이거나 고유명사일 경우 포함한다
            ma_selected.append(morph_lex)   
        elif morph_cat in FILTER_IN_POSES_AD1:
            ma_selected.append(morph_lex+'하다')
        elif morph_cat in FILTER_IN_POSES_AD2:
            ma_selected.append(morph_lex+'다')
        else:
            continue
    ma_results_selected.append(ma_selected)
    
print(ma_results_selected)
            
            
```

```python
###잠깐 딴 짓. 워드클라우드 만들기

from collections import Counter

# 빈도 높은 단어들만 가져오기
words_count=Counter()
for ma_selected in ma_results_selected:
    words_count.update(ma_selected)

# print(words_count)    
    
words_most = []
words_most_onlyword=[]
for word,count in words_count.most_common(500):
    words_most.append((word,count))
    words_most_onlyword.append(word)
    
print(words_most)  
```

```python
%matplotlib inline

from wordcloud import WordCloud

tmp = dict(words_most)
wordcloud=WordCloud(font_path="c:/Windows/Fonts/malgun.ttf", relative_scaling=0.3, 
                    background_color="white", max_words=50,
                   font_step=10).generate_from_frequencies(tmp)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('휴대폰_워드클라우드(white).png')
```

```python
# 단어 관계 행렬 만들기

from itertools import combinations

word_cooc_mat=Counter()

for word1, word2 in combinations(words_most,2):
    word_cooc_mat[(word1,word2)]+=1
    

    
```

```python
# 단어 관계 행렬 쓰기 
from collections import Counter
from itertools import combinations

word_cooc_mat=Counter()

for line in ma_results_selected:
    for word1, word2 in combinations(line,2):
        word_cooc_mat[(word1,word2)]+=1


    
```

```python
#탐색용. 삭제가능

# print(word_cooc_mat.items())
for a,b in word_cooc_mat.items():
    print(a,b)
    break
# print(word_cooc_mat)
```

```python
# 단어관계 행렬  counter 형태 --> 리스트 내 튜플 형태로 수정 --> 정렬하기

word_coocs=[]
for words, count in word_cooc_mat.items():
    word_coocs.append((words[0],words[1],count))
# print(word_coocs)    
        
sorted_word_coocs = sorted(word_coocs, key=lambda x: x[2], reverse=True)   # 정렬하기
print(sorted_word_coocs)
```

```python
#그래프 그리기
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#탐색용. 삭제가능

sorted_word_coocs[:40]
```

```python
G= nx.Graph()

for word1, word2, count in sorted_word_coocs[:40]:    #상위 n개로만 그림 그리기
    G.add_edge(word1, word2, weight=count)
    
T = nx.minimum_spanning_tree(G)

  
```

```python
T.nodes()
```

```python
T.
```

```python
T.edges()
```

```python
# 그래프 그리기

import sys

nodes = nx.nodes(T)
degrees = nx.degree(T)

node_size = []

for node in nodes:
    ns = degrees[node]*1000
    
    node_size.append(ns)
    
if sys.platform in ["win32", "win64"]:
    font_name = "malgun gothic"
elif sys.platform == "darwin":
    fornt_name = "AppleGothic"
    
    
plt.figure(figsize=(30,20))    
nx.draw(T,
       pos=nx.fruchterman_reingold_layout(G, k=0.5),
       node_size=node_size,
       node_color="#FFE27F",
       font_family=font_name,
       label_pos=0.5, #0=head, 0.5=center, 1=tail
        with_labels=True,
        font_size=20 )
plt.savefig("휴대폰댓글graph_단어추가.png")
plt.axis("off")

plt.show()
```
