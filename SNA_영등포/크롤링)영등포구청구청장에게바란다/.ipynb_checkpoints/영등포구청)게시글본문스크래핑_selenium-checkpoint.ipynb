```python
def get_html(url):
    '''
    url의 html 정보를 가져온다
    '''
    try:
        user_agent="Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_4) " +\
        "AppleWebKit/537.36 (KHTML, like Gecko) " + \
        "Chrome/37.0.2062.94 Safari/537.36"
        headers = {"User-Agent": user_agent}
        response=requests.get(url, headers=headers) 
        return response
    except requests.exceptions.ConnectionError:
        return 'Connection Failed!'
```

```python
# 게시글에 접속하여, 정보를 읽어온다

def get_texts(letter_url):
    html = get_html(letter_url)
    html = html.text
    soup = BeautifulSoup(html, 'lxml')
#     datas = soup.select('div.bbs_view_txt') 
    datas = soup.select('body > div.contents > div > div.row > div')
#     datas = soup.find('div', class_='bbs_view_txt')
    print(datas)
#     for data in datas:
#         print(data.text)
#     print(data)
    
#     print(soup)
    
    
url_test = 'http://www.ydp.go.kr/leader/minwon.do?cfgIdx=3&pageNm=B0101&page=4&op=view&idxId=109031'
# http://www.ydp.go.kr/leader/minwon.do?cfgIdx=3&pageNm=B0101&page=4&op=view&idxId=109031
get_texts(url_test)

```

```python
from selenium import webdriver
```

```python
# 셀레니움 이용하여 접속, html을 가져온다

# driver = webdriver.Chrome('C:/Users/namso_000/Downloads/chromedriver')

def get_html_by_selenium(url):
#     driver = webdriver.Chrome('C:/Users/namso_000/Downloads/chromedriver')
    driver.get(url)
    driver.implicitly_wait(3)
    time.sleep(random.uniform(0.3,0.7)) # 추가로 시간
    html = driver.page_source
    
    return html

```

```python
def get_texts(letter_url):
    html = get_html_by_selenium(letter_url)
    soup = BeautifulSoup(html, 'lxml')
    datas = soup.select('div.bbs_view_txt')[0]
    texts=''
    texts += datas.text.strip().replace('\n',' ')
#     print(texts)
    return(texts)
    
# get_texts(url_test)
```

```python
# 링크(분할) --> 텍스트 가져와서 저장하기
START = 401
END = 660

OUTPUT = './data/크롤링결과_{}to{}.txt'.format(START,END)

letters = []
driver = webdriver.Chrome('C:/Users/namso_000/Downloads/chromedriver')

for num in range(START, END+1):

    text = [get_texts(links[num-1])]
    letters.append(text)
    print('{}번째 게시글의 본문 저장중입니다'.format(num))
#     print(text[:30])

driver.close



with open(OUTPUT, 'w', encoding='utf-8') as texts:
    for letter in letters:
        texts.write(letter[0])
        texts.write('\n')
    print('저장완료')
```
