```python
from konlpy.tag import Komoran
```

```python
#데이터 불러오기.  한줄에 댓글 하나 → 리스트로 댓글별 저장
INPUT_FILE = './data/크롤링결과_total.txt'

datas=[]
with open(INPUT_FILE, "r", encoding='utf-8') as input_file:
    for line in input_file:
        if len(line) > 3:      # 줄간격 및 공백은 삭제
#             print(len(line))
            datas.append(line.strip())
#     data=input_file.read()

print(len(datas))
print(datas[:10])        
```

```python
# 댓글별 형태소 분석하기

ma_results = []
komoran = Komoran()

for line in datas:
    ma_by_line=komoran.pos(line)
    ma_results.append(ma_by_line)

print(ma_results[:2])
```

```python
#형태소분석 결과 저장하기
OUTPUT_ma_results = './data_구청장/형태소분석결과.txt'

with open(OUTPUT_ma_results, "w", encoding='utf-8') as output:
    for person in ma_results:
        for word in person:
            output.write(word[0]+',')
        output.writelines('\n')
    print('저장완료')
```

```python
# 댓글별 원하는 품사만 저장하기
# STOPWORD 제외하기

# FILTER_IN_POSES=['NNG','NNP']   
FILTER_IN_POSES=['NNG','NNP', 'XR']   
# FILTER_IN_POSES_AD1=['XR']     #어근.~하다 ex) 유쾌  
# FILTER_IN_POSES_AD2=['VA']      #형용사. ~다 ex) 쉽, 어렵
FILTER_IN_POSES_AD1=[]    #명사만 테스트
FILTER_IN_POSES_AD2=[]    #명사만 테스트

#NNG 일반명사  NNP 고유명사 XR 어근 (~하다) VA 형용사(~다)
STOPWORD=['구청','영등포구청','영등포구']   #분석에서 제외할 단어 추가 필요

ma_results_selected=[]

for ma_result in ma_results:
    ma_selected=[]
    for morph_lex, morph_cat in ma_result:
        if morph_lex in STOPWORD:
            continue
        elif morph_cat in FILTER_IN_POSES:   # 일반명사이거나 고유명사일 경우 포함한다
            ma_selected.append(morph_lex)   
        elif morph_cat in FILTER_IN_POSES_AD1:
            ma_selected.append(morph_lex+'하다')
        elif morph_cat in FILTER_IN_POSES_AD2:
            ma_selected.append(morph_lex+'다')
        else:
            continue
#         ma_selected=list(set(ma_selected))   #중복은 제외: 한 사람이 동일한 단어 여러번 이야기 했을 경우 제외함
    ma_results_selected.append(ma_selected)
    
print(ma_results_selected[:2])
            
            
```

```python
#형태소분석 결과 저장하기
# OUTPUT_ma_results = './data_구청장/형태소분석결과_명사어근형용사만.txt'
OUTPUT_ma_results = './data_구청장/형태소분석결과_명사만_중복미제외.txt'

with open(OUTPUT_ma_results, "w", encoding='utf-8') as output:
    for person in ma_results_selected:
            for word in person:
                output.write(word+',')
            output.writelines('\n')
            
    print('저장완료')
```

```python
###잠깐 딴 짓. 워드클라우드 만들기

from collections import Counter

# 빈도 높은 단어들만 가져오기
words_count=Counter()
for ma_selected in ma_results_selected:
    words_count.update(ma_selected)

# print(words_count)    
    
words_most = []
words_most_onlyword=[]
for word,count in words_count.most_common(500):
    words_most.append((word,count))
    words_most_onlyword.append(word)
    
print(words_most[:100])  
```

```python
# 단어 관계 행렬 쓰기 
from collections import Counter
from itertools import combinations

word_cooc_mat=Counter()

for line in ma_results_selected:
    for word1, word2 in combinations(line,2):
        if word1 == word2 :   #동일한 단어간의 벡터는 계산하지 않음
            continue
        elif word_cooc_mat[(word2,word1)]>=1:
            word_cooc_mat[(word2,word1)]+=1
        else:
            word_cooc_mat[(word1,word2)]+=1

    
```

```python
print(word_cooc_mat)
```

```python
#탐색용. 삭제가능

# print(word_cooc_mat.items())
for a,b in word_cooc_mat.items():
    print(a,b)
    break
# print(word_cooc_mat)
```

```python
# 단어관계 행렬  counter 형태 --> 리스트 내 튜플 형태로 수정 --> 정렬하기

word_coocs=[]
for words, count in word_cooc_mat.items():
    word_coocs.append((words[0],words[1],count))
# print(word_coocs)    
        
sorted_word_coocs = sorted(word_coocs, key=lambda x: x[2], reverse=True)   # 정렬하기
print(sorted_word_coocs)
```

```python
#그래프 그리기
import networkx as nx
import matplotlib.pyplot as plt
import sys
```

```python
#탐색용. 삭제가능

sorted_word_coocs[:40]
```

```python
# 그래프 그리기

G= nx.Graph()

for word1, word2, count in sorted_word_coocs[:100]:    #상위 n개로만 그림 그리기
    G.add_edge(word1, word2, weight=count)
    
T = nx.minimum_spanning_tree(G)


nodes = nx.nodes(T)
degrees = nx.degree(T)

node_size = []

for node in nodes:
    ns = degrees[node]*100
    
    node_size.append(ns)
    
if sys.platform in ["win32", "win64"]:
    font_name = "malgun gothic"
elif sys.platform == "darwin":
    fornt_name = "AppleGothic"
    
    
plt.figure(figsize=(30,20))    
nx.draw(T,
       pos=nx.fruchterman_reingold_layout(G, k=0.5),
       node_size=node_size,
       node_color="#FFE27F",
       font_family=font_name,
       label_pos=1, #0=head, 0.5=center, 1=tail
        with_labels=True,
        font_size=30 )
plt.savefig("./data_구청장/명사만_중복포함_관계상위100(1).png")
plt.axis("off")

plt.show()
```
